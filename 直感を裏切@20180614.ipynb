>生日悖论

>在某个班级里一共有23 名学生。不考虑双胞胎、闰年等特殊情况，在这些男孩女孩中间，有2 个人生日相同的概率是多少？

>如果在班级里发现某个同学和自己生日是同一天，一般人还是会感到些许惊讶吧。毕竟除了闰年以外，一年中有365 天，而偏偏两个人都出生在了同一天，就算是偶然，但要说没有一丝缘分的话似乎也说不过去。也就是说，在我们的直觉中，两人生日相同的概率应该是非常小的，这种缘分似乎并不常见。

>但是，数学与生活又开了个玩笑，你认为不太可能的缘分，实际发生的概率要高达**50.7%**。

```python
import pandas as pd
import numpy as np
```

```python
def factorial(n):
    for i in range(1 , n):
        n=n*i
    return n
```

```python
def P(x):
    p = (365**x - factorial(365) / factorial(365-x) ) / 365**x
    return p
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='white')
x = [i for i in range(100)]
y = [P(j) for j in x]
plt.plot(x,y)
plt.xlabel('num. of students')
plt.ylabel('probability')
plt.show()
```
